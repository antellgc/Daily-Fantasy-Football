# ESPN Data Scraping -- player projections

```python
week1_projections_url = "http://games.espn.com/ffl/tools/projections?&scoringPeriodId=1&seasonId=2017"
week2_projections_url = "http://games.espn.com/ffl/tools/projections?&scoringPeriodId=2&seasonId=2017"
week3_projections_url = "http://games.espn.com/ffl/tools/projections?&scoringPeriodId=3&seasonId=2017"
week4_projections_url = "http://games.espn.com/ffl/tools/projections?&scoringPeriodId=4&seasonId=2017"
```

```python
def get_projection_df(url_string):
    page = requests.get(url_string)
    soup = BeautifulSoup(page.content, 'html.parser')
    stats_table = soup.find('table', {'id': 'playertable_0'})
    row_stats = stats_table.findAll('tr',  {'class': ['playerTableBgRowSubhead','pncPlayerRow']})
    total_data = []
    for row in row_stats:
        data = [td.text.replace(u'\xa0',' ').encode('utf-8').strip() for td in row.find_all('td') if td.text]
        total_data.append(data)
    df = pd.DataFrame(total_data)
    df.rename(columns=df.iloc[0], inplace=True)
    df = df.reindex(df.index.drop(0))
    return df[['PLAYER, TEAM POS','PTS']]

projection_df = get_projection_df(week1_projections_url)
projection_df.head(10)
```
